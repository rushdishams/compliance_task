Load the original sentences and their augmentations

```python
import pandas as pd
import json
```

```python
# Read input file
with open('../data/augmentations.json') as data_file:    
    data = json.load(data_file)
```

Creating a dataframe. Each of its rows will contain augmentations paired with their respective original sentence. 
Each original sentence and augmentations will have unique ids. 

```python
# Generate rows containing (augmentation, original) sentence pairs
rows = [(oid, orig, aid, aug) for oid, orig in enumerate(data) for aid, aug in enumerate(data[orig])]
```

```python
# Create dataframe with the pairs
df = pd.DataFrame(rows, columns=['oid', 'orig', 'aid', 'aug'])
```

```python
# Peek
df.head()
```

Lowercasing the original and augmented sentences. This will be useful for preprocessing and also while vectorizing them.

```python
# Changing all text to lowercase
for col in ['orig', 'aug']:
    df[col + '_lower'] = df[col].str.lower()
```

Text Preprocssing. First, remove all punctuations and keep only alphanumerics.
Next, remove stopwords and finally stem the function words.

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Configure stopword remover and stemmer
stops = stopwords.words('english')
stemmer = PorterStemmer()

# Regex to remove punctuations and to keep alphanumerics only
NOPUNCT_REGEX = re.compile(r'[/(){}\[\]\|@,;\.!]')
ALNUM_REGEX = re.compile(r'[^a-zA-Z0-9  ]')

# Function for text preprocessing.
def clean_text(text):
    text = str(text)
    text = NOPUNCT_REGEX.sub(' ', text)
    text = ALNUM_REGEX.sub('', text)
    text = ' '.join([w for w in text.split() if w not in stops])
    text = ' '.join([stemmer.stem(w) for w in text.split()])
    return text
```

```python
# Apply regex filters, remove stops and stem function words
df['orig_cleaned'] = df['orig_lower'].apply(clean_text)
df['aug_cleaned'] = df['aug_lower'].apply(clean_text)
```

```python
# Peek
df.head()
```

```python
# get unique cid
# for each cid, send df to a function
# The function fits a vectorizer on a_cleaned, and transforms
# The function transforms the vectorizer on c_cleaned[0]
# Gets cosine similarity for (c_cleaned, a_cleaned)
# Finally adds the sim score for the rows
```

Create similarity filter. For each original sentence, we send the associated data to a function. The function fits transforms a count vectorizer on clean augmentations and then simply transforms the original sentence. This provides us count vectors for the original sentence and all of its augmentations. Using these vectors we measure the similarity of the augmentations given the original sentence. 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Compute similarity between the augmenations and their original sentence
def get_sim_score(df):
    vectorizer = CountVectorizer()
    aug_vec = vectorizer.fit_transform(df['aug_cleaned'])
    orig_vec = vectorizer.transform(df.iloc[:1]['orig_cleaned'])
    df['sim_scores'] = cosine_similarity(orig_vec, aug_vec).flatten()
    return df
```

```python
# List of original sentence ids
oid_list = df['oid'].unique()
# Dataframe to contain similarity for all (augmentated, original) sentence pairs
final_df = pd.DataFrame()
# Vertically tack similarity score for the pairs for each original sentence
for oid in oid_list:
    final_df = pd.concat([final_df, get_sim_score(df[df['oid'] == oid])], axis=0)
```

```python
# Checking if the shapes are equal. 
# final_df will have one extra column (similarity score)
final_df.shape, df.shape
```

Grammar Check Filter. Each augmentation is checked with an off-the-shelf library. The library provides total grammatical mistakes present in them. 

```python
import language_tool_python as lt

# Initializing for grammar checking
tool = lt.LanguageTool('en')

# Total grammar mistakes in augmentations
def get_grammar_mistakes(s):
    return len(tool.check(s))
```

```python
# Get grammar mistakes for each augmentation
final_df['grammar_mistakes'] = final_df['aug'].apply(lambda x: get_grammar_mistakes(x))
```

```python
# Histogram of grammar mistakes show that the distribution is long-tailed
# This means we can use an Inter-Quartile Range (IQR) based noise removal
# technique to find a threshold. Any augmentation having grammar mistakes
# higher than this threshold will fail. We will not consider the lower range
# of the threshold since that indicates almost no grammar mistakes or good augmentations
final_df['grammar_mistakes'].hist()
```

```python
# Get 75th and 25th percentile of grammar mistakes
grammar_q3 = final_df['grammar_mistakes'].quantile(0.75)
grammar_q1 = final_df['grammar_mistakes'].quantile(0.25)
# Calculate inter-quartile range and then the upper bound
grammar_iqr = grammar_q3 - grammar_q1
grammar_cutoff = grammar_q3 + 1.5 * grammar_iqr
print(grammar_cutoff)
```

```python
# The similarity scores distribution is rather even.
# Therefore, an Inter-Quartile Range (IQR) based noise
# detection method cannot be used to find a threshold.
final_df['sim_scores'].hist()
```

```python
import numpy as np
SIM_CUTOFF = 0.5

final_df['sim_filter'] = np.where(final_df['sim_scores'] >= SIM_CUTOFF, True, False)
final_df['grammar_filter'] = np.where(final_df['grammar_mistakes'] >= grammar_cutoff, True, False)
```

```python
# Generate messages to understand the reason of pass/fail easy
def create_check_msg(row):
    msg = (str(round(row['sim_scores'] * 100, 1)) + '% similarity with ' +  
           str(round(row['grammar_mistakes'], 3)) + ' grammar mistake(s)')
    return msg
```

```python
final_df['checks'] = final_df.apply(create_check_msg , axis=1)
```

Formatting results for output

```python
# If an augmenation passes all filters, then it passed.
# If it fails in any, then it failed.
passed = final_df[(final_df['sim_filter'] == False) & (final_df['grammar_filter'] == False)]
failed = final_df[(final_df['sim_filter'] == True) | (final_df['grammar_filter'] == True)]
```

```python
passed.shape, failed.shape
```

```python
# Creating the json payload (output)
payload = dict()
payload['passed'] = passed[['aug', 'checks']].to_json(orient='records')
payload['failed'] = passed[['aug', 'checks']].to_json(orient='records')
```
